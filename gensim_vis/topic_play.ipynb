## В библиотеке gensim есть встроенная интерактивная визаулизация расстояния между темами. Можно использовать ее для того, чтобы оценить насколько пересекаются между собой полученные темы.

Применим ее для обеих моделей

```python
!pip3 install pyLDAvis
import pyLDAvis
import pyLDAvis.gensim_models as gensimvis
```

```python
import pandas as pd              # Пакет для работы с таблицами
import numpy as np               # Пакет для работы с векторами и матрицами
import pickle                    # Пакет для подгрузки данных специфического для питона формата
```

```python
pyLDAvis.enable_notebook()
```

```python
with open('/content/lda_20_base', 'rb') as f:
    ldamodel_base = pickle.load(f)
with open('/content/lda_20_base.state', 'rb') as f:
    state=pickle.load(f)
expElogbeta=np.load('/content/lda_20_base.expElogbeta.npy')
ldamodel_base.expElogbeta =expElogbeta
with open('/content/lda_20_base.state', 'rb') as f:
    state=pickle.load(f)
ldamodel_base.state=state    
with open('/content/corpus_base', 'rb') as f:
    corpus_base = pickle.load(f)
with open('/content/dictionary_base', 'rb') as f:
    dictionary_base = pickle.load(f) 
```

```python
vis_base = gensimvis.prepare(ldamodel_base, corpus_base, dictionary_base)
```
